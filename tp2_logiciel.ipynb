```python
import tkinter as tk
from PIL import Image,ImageTk

def routerCreate(event):
    
    global tags
    
    tags = cnv.create_image(event.widget.winfo_rootx(), event.widget.winfo_rooty()-120, anchor="nw", 
                     image=routerImage)

def switchCreate(event):
    cnv.create_image(event.widget.winfo_rootx(), event.widget.winfo_rooty()-120, anchor="nw", 
                     image=switchImage)
    
def phoneCreate(event):
    cnv.create_image(event.widget.winfo_rootx(), event.widget.winfo_rooty()-120, anchor="nw", 
                     image=phoneImage)
    
def pcCreate(event):
    cnv.create_image(event.widget.winfo_rootx(), event.widget.winfo_rooty()-120, anchor="nw", 
                     image=pcImage)
    
def lineCreate(event):
    cnv.create_image(event.widget.winfo_rootx(), event.widget.winfo_rooty()-120, anchor="nw", 
                     image=lineImage)
    
def motion(event):
    #Le canva est un widget et il gère lui même ses objets. Ils possèdent tous un numéro 
    #unique. C'est le tag ou id
    #On récupère le tag de l'élément le plus proche de la souris
    id = cnv.find_closest(event.x, event.y)[0]
    #On peut déplacer l'objet grâce à moveto et son tag/id
    cnv.moveto(id, event.x, event.y)


#Création fenêtre principale et son titre
root = tk.Tk()
root.title("Cisso tracer")

# 0      1      2
#Canva Canva Canva    0
#Toolbar Canva Canva  1
#Canva Canva Canva    2

#On indique le poids de chacunes des lignes et colonnes 
#de la fenêtre principale pour les agencer correctements
root.grid_rowconfigure(0, weight=1)# 1er ligne
root.grid_rowconfigure(1, weight=1)
root.grid_rowconfigure(2, weight=1)
root.grid_columnconfigure(0, weight=0)# 1er colonne
root.grid_columnconfigure(1, weight=1)

#Frame toolbar
toolbar = tk.Frame(root)

##Boutons images
    
img= Image.open("router.png")
img= img.resize((200,200))
routerImage=ImageTk.PhotoImage(img)
routerButton= tk.Button(toolbar, image=routerImage)
routerButton.bind("<ButtonRelease-1>", routerCreate)
routerButton.grid(row=0)
root.update()

img= Image.open("switch.png")
img= img.resize((200,200))
switchImage=ImageTk.PhotoImage(img)
switchButton= tk.Button(toolbar, image=switchImage)
switchButton.bind("<Button-1>", switchCreate)
switchButton.grid(row=1)

img= Image.open("pc.png")
img= img.resize((200,200))
pcImage=ImageTk.PhotoImage(img)
pcButton= tk.Button(toolbar, image=pcImage)
pcButton.bind("<Button-1>", pcCreate)
pcButton.grid(row=3)

img= Image.open("phone.png")
img= img.resize((200,200))
phoneImage=ImageTk.PhotoImage(img)
phoneButton= tk.Button(toolbar, image=phoneImage)
phoneButton.bind("<Button-1>", phoneCreate)
phoneButton.grid(row=2)

img= Image.open("line.png")
img= img.resize((200,200))
lineImage=ImageTk.PhotoImage(img)
lineButton= tk.Button(toolbar, image=lineImage)
lineButton.bind("<Button-1>", lineCreate)
lineButton.grid(row=4)

toolbar.grid(row=1, column=0)

#Canva où seront positionnés les traits et les items
cnv = tk.Canvas(root, bg="white")
cnv.grid(row=0, column=1, columnspan=3, rowspan=3, sticky="nswe")# nswe permet de prendre tout l'espace

cnv.bind("<B1-Motion>", motion)

#Boucle principale
root.mainloop()
```
